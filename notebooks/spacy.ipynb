```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
text ="I work in Apple which provide me good salary.I eat apple in US with elon musk."
doc=nlp(text)
for env in doc.ents:
    print(f'TEXT:{env.text}  LABEL: {env.label_}')
```

```python
text ="I work in Apple which provide me good salary.I eat apple in kathmandu."
doc=nlp(text)
for env in doc:
    print(f'TEXT:{env.text},  POS tag: {env.pos_}')
```
