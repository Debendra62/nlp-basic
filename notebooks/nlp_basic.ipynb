```python
test_para = '''School is the place where we learn to read and write. 
It is the most crucial place for a student, and it helps us to learn new things. 
The teachers are always helpful and teach us important things in life. 
We must always be regular to school as missing classes can lead to problems during exams. 
Schools teach us how to be consistent, punctual, and obedient. 
It also makes us better human beings so that we can treat our elders with respect. 
Most of what we learn is a result of the learning imparted by our teachers.
'''
```

## Tokenization

```python
# whitespace word tokenize normal(without using package)
test = test_para.replace("\n","")
test = test.split(" ")
test
```

```python
#  sentence tokenize
sentences = test_para.replace("\n","")
sentences = sentences.split(". ")
sentences
```

```python
# word tokenize package
from nltk.tokenize import word_tokenize
word_tokenize(test_para)
```

```python
# sentence tokenize package
from nltk.tokenize import sent_tokenize
sent_tokenize(test_para)
```

```python
# Stop Words

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
without_stopwords = [word for word in test if word not in stop_words]
para_without_stopwords = " ".join(without_stopwords)
print("Without Stop Words: ","\n", para_without_stopwords, "\n")
print("With Stop Words: ","\n", test_para, "\n")
```
